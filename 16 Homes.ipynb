```python
#define our plot method
def plot_state(state):
    print(state.shape)
    state = state.set_index('ZipCity')
    state_tr = state.transpose()
    county_list = state_tr.columns.values.tolist()
    
    fig = pp.figure()
    
    ax = fig.add_subplot(111)
    ax.set_xlabel('Year-Month')
    ax.set_ylabel('Median Value Per Square Foot')
    
    for name in county_list:
        state_tr[name].plot()
```

```python
plot_state(dc_df)
plot_state(va_df)
plot_state(md_df)
plot_state(wv_df)
```
