```python
#17 web scraping

import requests
from bs4 import BeautifulSoup
```

```python
#load a page
r = requests.get("https://www.yellowpages.com/search?search_terms=Coffee&geo_location_terms=Washington%2C+DC")
```

```python
#display content of r
r.content
```

```python
#declare a soup variable and read in the content
soup = BeautifulSoup(r.content, 'lxml')
```

```python
#beautify the content
print(soup.prettify())
```

```python
#list all the anchor tags ("a" tags)
soup.find_all("a")
```

```python
soup.find_all("h1")
```

```python
#show all the links
for link in soup.find_all("a"):
    print(link)
```

```python
#show all href
for link in soup.find_all("a"):
    print(link.get("href"))
```

```python
#print all the text
for link in soup.find_all("a"):
    print(link.text)
```

```python
#print all the text and the href
for link in soup.find_all("a"):
    print(link.text, link.get("href"))
```

```python
#get all the links and display in proper format
for link in soup.find_all("a"):
    print("<a href='%s'>%s</a>" %(link.get("href"), link.text))
```

```python
g_data = soup.find_all("div", attrs={"class":"info"})
g_data
```

```python
#cleanup the result a bit
for item in g_data:
    print(item.text)
```

```python
#get items as a list
for item in g_data:
    print(item.contents)
```

```python
#get the first two items of a list
items_list = []
for item in g_data:
    items_list.append(item)
items_list[0:2]
```

```python
#print the 0th element content and text
for item in g_data:
    print(item.contents[0])

for item in g_data:
    print(item.contents[0].text)
```

```python
#get the name and address of the places
for item in g_data:
    print(item.contents[0].text)
    print(item.contents[1].text)
```

```python
#get more specific using the class and build a list
for item in g_data:
    print(item.contents[0].find_all("a", attrs={"class":"business-name"}))
```

```python
#make a pretty list of locations, names, and addresses
for item in g_data:
    print(item.contents[0].find_all("a", {"class":"business-name"})[0].text)
    print(item.contents[1].find_all("p", {"class":"adr"})[0].text)
    print(item.contents[1].find_all("div", {"itemprop":"telephone"})[0].text)
    print()
```
