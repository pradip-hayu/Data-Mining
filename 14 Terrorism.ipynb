```python
#creating year column from the date column
#this will allow us to group the data by year to get annual stats
df['Year'] = df['Date'].dt.year
```

```python
#creating a count variable for recording the number of terrorist attacks 
#when grouping by another variable
df['Number of Terrorist Attacks'] = 1
```

```python
#creating dataframe DFA that contains onlu terrorsit attacks from known groups
dfa = df[df.Perpetrator != "Unknown"]
dfa = dfa[dfa.Perpetrator != "Other"]
```

```python
dfa.head()
```

```python
#groupng the original dataframe by YEar and then summing other columns for each year
ts = df.groupby('Year').sum()
```

```python
ts.head()
```

```python
#creating a time series that displays Injuries, Fatalities, and Number of Terrorist Attacks
ax = ts.plot(figsize=(20, 10), fontsize=13)
plt.legend(loc='best', fontsize=13)
ax.set_xlabel('Year', fontsize=13)
ax.set_ylabel('Number', fontsize=13)
```

```python
#creata a monthly stats for original dataframe
df['Month'] = [df.to_period('M') for df in df['Date']]
df['Month'].head()
```

```python
#create monthly stats
monthly = df.groupby(['Month']).sum()
monthly = monthly.drop('Year', axis=1)
monthly.head()
```

```python
#one problem that we might come across is if there are any missing months in the data
#let's look at the total months
monthly
```

```python
#the total number of months is
((2009-1968)+1)*12
```

```python
#creating a time series that displays Injuries, Fatalities, and Number of Terrorist Attacks
#monthly from 1968-2009
ax = monthly.plot(figsize=(20, 10), fontsize=13)
plt.legend(loc='best', fontsize=13)
ax.set_xlabel('Year', fontsize=13)
ax.set_ylabel('Number', fontsize=13)
```

```python
#we do see some spikes in the monthly data so let's filter them
dat = monthly.ix[(monthly['Fatalities'] >=2000) | (monthly['Injuries'] >= 2000)]
dat.plot(kind='bar', figsize=(10, 6), title='Top Deadly Months in Terror Attacks')
```

```python
#what about most deadly terrorist attacks?
inter = df.ix[(df['Injuries'] >=1500) | (df['Fatalities']>=1500)]
inter = inter.drop('Month', axis=1)
inter
```

```python
#calculating the percentage of fatalities as total number of casualties
monthly['Fatality Percent'] = (monthly['Fatalities']/(monthly['Fatalities']+monthly['Injuries']))*100
```

```python
monthly['Fatality Percent'].head(10)
```

```python
#let's get rid of the NaNs
monthly['Fatality Percent'] = monthly['Fatality Percent'].fillna(0)
monthly.head(10)
```

```python
#line graoh of the Fataltiy Percent
ax = monthly['Fatality Percent'].plot(figsize=(20, 10), fontsize=13)
plt.legend(loc='best', fontsize=13)
ax.set_xlabel('Year', fontsize=13)
ax.set_ylabel('Percent', fontsize=13)
```

```python
#calculating Fatality Percent for each attack as we did above but for the original dataframe
df['Fatality Percent'] = (df['Fatalities']/(df['Fatalities']+df['Injuries']))*100
df['Fatality Percent'] = df['Fatality Percent'].fillna(0)
df.head()
```

```python
#grouping by year to get the yearly stats for original dataframe
total = df.groupby('Year')
total = total.mean()
total
```

```python
#plotting the yearly fatality percent
ax = total['Fatality Percent'].plot(figsize=(20, 10), fontsize=13)
plt.legend(loc='best', fontsize=13)
ax.set_xlabel('Year', fontsize=13)
ax.set_ylabel('Percent', fontsize=13)
```

```python
#create a new dataframe called dayint
dayint = df

#create a new column in dayint with unique days of year for each terrorist attack
dayint['Day'] = dayint['Date'].dt.dayofyear
dayint['Day'].head(10)
```

```python
#group by each unique day of the year and sum remaining variables
dayint = dayint.groupby('Day').sum()
dayint = dayint.drop('Year', axis=1)
```

```python
dayint
```

```python
#plot attacks by unique day of the year
ax = dayint['Number of Terrorist Attacks'].plot(figsize=(20,10), fontsize=13)
ax.set_xlabel('Unique Day of the Year')
ax.set_ylabel('Number of Terrorist Attacks')
```

```python
#using the Contains function we can find any descriptions of a terror attack that includes
#words like embassy or consulate or mission
embatt = pd.DataFrame(df.Description.str.contains('embassy|consulate|mission', 
                                                  case=False).value_counts())
embatt
```

```python
#rename the heading
embatt = embatt.rename(columns={'Description':'Number of Attacks'})
embatt
```

```python
#renaming the index of the dataframe
embatt.index = ["Non-Diplomatic Targets", "Diplomatic Targets"]
embatt
```

```python
from matplotlib.pyplot import *

#plot the embatt dataframe
fig, ax = subplots()
embatt.plot(kind='bar', ax=ax, rot=0)
```

```python
#using the DFA dataframe to find the top 15 terrorist groups in terms of Number of Attacks
perp = dfa.groupby('Perpetrator').sum()
terr = perp.sort_values(by='Number of Terrorist Attacks', ascending=False).head(15)
```

```python
terr
```

```python
#plot this information
pd.Series(terr['Number of Terrorist Attacks']).plot(kind='barh', 
                                                    title='Number of Terrorist Attacks by Groups')
```

```python
#use the DFA dataframe to find the top 15 terrorist groups in terms of Fatalities
fatal = perp.sort_values(by='Fatalities', ascending=False).head(15)
pd.Series(fatal['Fatalities']).plot(kind='barh', title='Fatalities by Terrorist Groups')
```

```python
#calculate the number of fatalities from Al Qaeda and pass to variable data
data = pd.Series(fatal.ix[['Al Qaeda'],['Fatalities']])
data
```

```python
#calculate the percentage of fatalities from Al Qaeda compared to other terrorist groups
#including the Unknown and Other groups; this is a pie chart

plt.rcParams['font.size'] = 12
plt.figure(figsize=(5,5))

sizes = [data, (df['Fatalities'].sum()-data)]
labels = ['Al Qaeda', 'Other Terrorist Groups']
colors = ['Green', 'Red']
figure = (10, 10)
plt.pie(sizes, colors=colors, shadow=True, startangle=140, autopct='%1.1f%%')
plt.legend(labels, bbox_to_anchor=(0.05, 0.8), fontsize=12)
plt.axis('equal')
plt.show()
```
